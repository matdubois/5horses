In the data exploration, I saw that in census_data 2010 there was one column, "id2," without any values in it. I chose to delete the column from the dataset to increase its readability by using the drop function. In the same dataset, the column "target geo id2" has codes for all of the counties in the United States. The CDC_ALL file has the same codes in the column FIPS codes. I used the rename function so that the column name in both datasets match. This will allow us to merge them using the pd.merge function. 

```python
# Checking to see if the columns that I want to rename have a hidden format
census_data.columns
```

```python
#removing columns with duplicate information and with no values
#drop([census_data.columns[1] used this code because as can be seen in the columns function there are two columns 
#that have the name geographic area
census_data_clean = census_data.drop([census_data.columns[1], census_data.columns[5]], axis="columns")
census_data_clean               
```

```python
# renaming Target geo ID2 to FIPS Codes so that we are able to merge these dataframes using pd.merge 
#renaming geographic area1 for aesthetic reasons 
census_data_clean= census_data_clean.rename(columns = {"Target Geo Id2": "FIPS Codes", "Geographic area.1":"County" }) 
census_data_clean
```

```python
# removing rhe NA value from the dataset to allow for exploration 
census_data_clean = census_data_clean.fillna(0)
census_data_clean
```

```python
census_data_clean.columns
```

```python
#changed the format iN FIPS codes to match the format that FIPS codes is in the cdc_diabetes

census_data_clean["FIPS Codes"].astype(int)

census_data_clean["FIPS Codes"] = census_data_clean["FIPS Codes"].astype(int)

#census_data_clean
```

```python
#removing columns with data that we don't need. 

cdc_diabetes_clean = cdc_diabetes_data.drop(["lower confidence limit_2010", "upper confidence limit_2010", "age-adjusted percent_2010",
                        "age-adjusted lower confidence limit_2010", "age-adjusted upper confidence limit_2010",
                       "lower confidence limit_2013", "upper confidence limit_2013", "age-adjusted percent_2013"
                       , "age-adjusted lower confidence limit_2013", "age-adjusted upper confidence limit_2013"], axis=1)
cdc_diabetes_clean.head()
```

```python
diabetes_population_data_df = diabetes_population_data_df.drop(["Id", "Target Geo Id", "Geography", "Area in square miles - Total area", 
                                  "Area in square miles - Water area", "Area in square miles - Land area", "County_y"], axis=1)
diabetes_population_data_df
```

```python
diabetes_population_data_df = diabetes_population_data_df.rename(columns = {"County_x": "County", 
                                                                            "number_2010": "Number of People with Diabetes 2010", 
                                                                            "percent_2010":"Percentage of People with Diabetes 2010",
                                                                            "number_2013": "Number of People with Diabetes 2013",
                                                                            "percent_2013": "Percentage of People with Diabetes 2013"})

diabetes_population_data_df
```
